```python
'''**2. Bank Account Management:**

- Create classes for `BankAccount`, `Customer`, and `Transaction`.
- Implement methods to deposit, withdraw, and transfer funds.
- Generate account statements and track transaction history.
- Implement security features like password protection and access control.'''
class Transaction:
    def __init__(self, amount, transaction_type):
        self.amount = amount
        self.transaction_type = transaction_type

class BankAccount:
    def __init__(self, account_number, balance=0, password=None):
        self.account_number = account_number
        self.balance = balance
        self.password = password
        self.transaction_history = []

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(Transaction(amount, 'Deposit'))

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.transaction_history.append(Transaction(amount, 'Withdrawal'))
        else:
            print("Insufficient balance")

    def transfer(self, amount, recipient_account):
        if self.balance >= amount:
            self.balance -= amount
            recipient_account.deposit(amount)
            self.transaction_history.append(Transaction(amount, 'Transfer to ' + recipient_account.account_number))
        else:
            print("Insufficient balance")

    def generate_statement(self):
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance}")
        print("Transaction History:")
        for transaction in self.transaction_history:
            print(f"{transaction.transaction_type}: {transaction.amount}")

class Customer:
    def __init__(self, name, bank_account):
        self.name = name
        self.bank_account = bank_account

# Example Usage:
# Create bank accounts
account1 = BankAccount("123456", password="password1")
account2 = BankAccount("789012", password="password2")

# Create customers
customer1 = Customer("Alice", account1)
customer2 = Customer("Bob", account2)

# Deposit, withdraw, and transfer funds
account1.deposit(1000)
account1.withdraw(500)
account1.transfer(200, account2)

# Generate account statements
account1.generate_statement()
account2.generate_statement()
```

```python
'''**3. Library Management System:**

- Create classes for `Book`, `Member`, and `Library`.
- Implement methods to borrow, return, and renew books.
- Manage book catalog, track availability, and handle overdue fines.
- Allow members to search for books, reserve titles, and view their borrowing history.'''
class Book:
    def __init__(self, title, author, isbn, available=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = available

class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.available:
            self.borrowed_books.append(book)
            book.available = False
            print(f"{self.name} has borrowed '{book.title}'")
        else:
            print(f"'{book.title}' is not available for borrowing")

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.available = True
            print(f"{self.name} has returned '{book.title}'")
        else:
            print(f"{self.name} has not borrowed '{book.title}'")

    def renew_book(self, book):
        if book in self.borrowed_books:
            print(f"{self.name} has renewed '{book.title}'")
        else:
            print(f"{self.name} has not borrowed '{book.title}'")

class Library:
    def __init__(self):
        self.catalog = []

    def add_book(self, book):
        self.catalog.append(book)

    def search_book(self, title):
        for book in self.catalog:
            if title.lower() in book.title.lower():
                print(f"Book found - Title: {book.title}, Author: {book.author}, ISBN: {book.isbn}, Available: {book.available}")
                return
        print("Book not found")

    def display_catalog(self):
        print("Library Catalog:")
        for book in self.catalog:
            print(f"Title: {book.title}, Author: {book.author}, ISBN: {book.isbn}, Available: {book.available}")

# Example Usage:
# Create library
library = Library()

# Add books to the library
book1 = Book("Python Programming", "John Smith", "978-0134845623")
book2 = Book("Introduction to Machine Learning", "Alice Johnson", "978-0262039406")
library.add_book(book1)
library.add_book(book2)

# Display library catalog
library.display_catalog()

# Create member
member = Member("Alice", "M001")

# Borrow a book
member.borrow_book(book1)

# Try to borrow a book that is already borrowed
member2 = Member("Bob", "M002")
member2.borrow_book(book1)

# Return a book
member.return_book(book1)

# Renew a book
member.renew_book(book1)

# Search for a book
library.search_book("Python")
```

```python
'''**4. Student Grading System:**

- Create classes for `Student`, `Course`, and `Grade`.
- Calculate overall grades based on assignments, quizzes, and exams.
- Generate student transcripts and summary reports.
- Track student enrollment and course completion.'''
class Grade:
    def __init__(self, assignment=0, quiz=0, exam=0):
        self.assignment = assignment
        self.quiz = quiz
        self.exam = exam

    def calculate_grade(self):
        total = self.assignment + self.quiz + self.exam
        return total

class Course:
    def __init__(self, name, max_score):
        self.name = name
        self.max_score = max_score

class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.enrolled_courses = {}

    def enroll_course(self, course):
        if course.name not in self.enrolled_courses:
            self.enrolled_courses[course.name] = Grade()

    def submit_assignment(self, course_name, score):
        if course_name in self.enrolled_courses:
            self.enrolled_courses[course_name].assignment = score
        else:
            print(f"{self.name} is not enrolled in {course_name}")

    def take_quiz(self, course_name, score):
        if course_name in self.enrolled_courses:
            self.enrolled_courses[course_name].quiz = score
        else:
            print(f"{self.name} is not enrolled in {course_name}")

    def take_exam(self, course_name, score):
        if course_name in self.enrolled_courses:
            self.enrolled_courses[course_name].exam = score
        else:
            print(f"{self.name} is not enrolled in {course_name}")

    def generate_transcript(self):
        print(f"Transcript for {self.name}:")
        for course_name, grade in self.enrolled_courses.items():
            overall_grade = grade.calculate_grade()
            print(f"Course: {course_name}, Overall Grade: {overall_grade}/{grade.exam}")

    def generate_summary_report(self):
        print(f"Summary Report for {self.name}:")
        total_courses = len(self.enrolled_courses)
        print(f"Total Courses Enrolled: {total_courses}")
        for course_name, grade in self.enrolled_courses.items():
            overall_grade = grade.calculate_grade()
            print(f"Course: {course_name}, Overall Grade: {overall_grade}/{grade.exam}")

# Example Usage:
# Create courses
math_course = Course("Mathematics", max_score=100)
science_course = Course("Science", max_score=100)

# Create student
student1 = Student("Alice", "S001")

# Enroll in courses
student1.enroll_course(math_course)
student1.enroll_course(science_course)

# Submit assignments, take quizzes, and exams
student1.submit_assignment("Mathematics", 90)
student1.take_quiz("Mathematics", 80)
student1.take_exam("Mathematics", 85)

student1.submit_assignment("Science", 85)
student1.take_quiz("Science", 75)
student1.take_exam("Science", 80)

# Generate transcript and summary report
student1.generate_transcript()
student1.generate_summary_report()
```

```python
'''**5. Shopping Cart System:**

- Create classes for `Product`, `Cart`, and `Order`.
- Implement methods to add, remove, and update items in the cart.
- Calculate total cost, apply discounts, and handle shipping options.
- Process payments and generate order confirmations.'''
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Cart:
    def __init__(self):
        self.items = {}

    def add_item(self, product, quantity):
        if product in self.items:
            self.items[product] += quantity
        else:
            self.items[product] = quantity

    def remove_item(self, product, quantity):
        if product in self.items:
            if self.items[product] > quantity:
                self.items[product] -= quantity
            else:
                del self.items[product]

    def update_item_quantity(self, product, quantity):
        if product in self.items:
            self.items[product] = quantity

    def calculate_total_cost(self):
        total_cost = 0
        for product, quantity in self.items.items():
            total_cost += product.price * quantity
        return total_cost

class Order:
    def __init__(self, cart):
        self.cart = cart

    def process_payment(self):
        total_cost = self.cart.calculate_total_cost()
        # Here you would include code to process payment, like connecting to a payment gateway
        print(f"Payment processed successfully. Total amount: ${total_cost:.2f}")

    def generate_confirmation(self):
        print("Order Confirmation:")
        for product, quantity in self.cart.items.items():
            print(f"Product: {product.name}, Quantity: {quantity}")
        total_cost = self.cart.calculate_total_cost()
        print(f"Total Cost: ${total_cost:.2f}")

# Example Usage:
# Create products
product1 = Product("Laptop", 1000)
product2 = Product("Mouse", 20)

# Create a cart
cart = Cart()

# Add items to the cart
cart.add_item(product1, 2)
cart.add_item(product2, 3)

# Remove items from the cart
cart.remove_item(product1, 1)

# Update item quantity in the cart
cart.update_item_quantity(product2, 5)

# Calculate total cost
total_cost = cart.calculate_total_cost()
print(f"Total Cost: ${total_cost:.2f}")

# Create an order and process payment
order = Order(cart)
order.process_payment()

# Generate order confirmation
order.generate_confirmation()
```

```python
'''**6. Movie Ticket Booking System:**

- Create classes for `Movie`, `Show`, `Seat`, and `Ticket`.
- Display movie schedules, show timings, and available seats.
- Allow users to select seats and book tickets.
- Process payments and generate ticket confirmations.'''
class Movie:
    def __init__(self, title, duration):
        self.title = title
        self.duration = duration

class Show:
    def __init__(self, movie, timing, available_seats):
        self.movie = movie
        self.timing = timing
        self.available_seats = available_seats

    def display_show_details(self):
        print(f"Movie: {self.movie.title}")
        print(f"Timing: {self.timing}")
        print(f"Available Seats: {self.available_seats}")

class Seat:
    def __init__(self, row, number):
        self.row = row
        self.number = number

class Ticket:
    def __init__(self, show, seats):
        self.show = show
        self.seats = seats

    def calculate_total_cost(self):
        return len(self.seats) * 10  # Assuming each ticket costs $10

class BookingSystem:
    def __init__(self):
        self.shows = []

    def add_show(self, show):
        self.shows.append(show)

    def display_movie_schedule(self):
        print("Movie Schedule:")
        for i, show in enumerate(self.shows, start=1):
            print(f"Show {i}:")
            show.display_show_details()

    def book_tickets(self, show_index, selected_seats):
        if show_index < 1 or show_index > len(self.shows):
            print("Invalid show index")
            return

        show = self.shows[show_index - 1]
        for seat in selected_seats:
            if seat in show.available_seats:
                show.available_seats.remove(seat)
            else:
                print(f"Seat {seat.row}{seat.number} is not available")
                return

        ticket = Ticket(show, selected_seats)
        total_cost = ticket.calculate_total_cost()
        print("Tickets booked successfully!")
        print(f"Total cost: ${total_cost}")
        print("Enjoy the movie!")

# Example Usage:
# Create movies
movie1 = Movie("The Avengers", "2 hours")
movie2 = Movie("The Lion King", "1.5 hours")

# Create shows
show1 = Show(movie1, "12:00 PM", [Seat("A", 1), Seat("A", 2), Seat("B", 1)])
show2 = Show(movie2, "3:00 PM", [Seat("C", 3), Seat("C", 4), Seat("D", 1)])

# Create booking system
booking_system = BookingSystem()

# Add shows to the booking system
booking_system.add_show(show1)
booking_system.add_show(show2)

# Display movie schedule
booking_system.display_movie_schedule()

# Book tickets
booking_system.book_tickets(1, [Seat("A", 1), Seat("A", 2)])
```

```python
'''
**7. Fitness Tracker App:**

- Create classes for `User`, `Activity`, and `Goal`.
- Track daily steps, workouts, calorie intake, and sleep patterns.
- Set fitness goals, monitor progress, and provide motivational feedback.
- Generate personalized reports and recommendations.'''
class Activity:
    def __init__(self, date, steps=0, workouts=[], calorie_intake=0, sleep_hours=0):
        self.date = date
        self.steps = steps
        self.workouts = workouts
        self.calorie_intake = calorie_intake
        self.sleep_hours = sleep_hours

    def add_steps(self, steps):
        self.steps += steps

    def add_workout(self, workout):
        self.workouts.append(workout)

    def add_calorie_intake(self, calorie_intake):
        self.calorie_intake += calorie_intake

    def add_sleep_hours(self, sleep_hours):
        self.sleep_hours += sleep_hours

class Goal:
    def __init__(self, target_steps, target_workouts_per_week, target_calories_per_day, target_sleep_hours_per_night):
        self.target_steps = target_steps
        self.target_workouts_per_week = target_workouts_per_week
        self.target_calories_per_day = target_calories_per_day
        self.target_sleep_hours_per_night = target_sleep_hours_per_night

    def progress_report(self, activity):
        print("Progress Report:")
        print(f"Steps: {activity.steps}/{self.target_steps}")
        print(f"Workouts: {len(activity.workouts)}/{self.target_workouts_per_week} per week")
        print(f"Calorie Intake: {activity.calorie_intake}/{self.target_calories_per_day} per day")
        print(f"Sleep Hours: {activity.sleep_hours}/{self.target_sleep_hours_per_night} per night")

class User:
    def __init__(self, name, age, gender, weight, height):
        self.name = name
        self.age = age
        self.gender = gender
        self.weight = weight
        self.height = height
        self.activity_log = {}

    def log_activity(self, date, activity):
        self.activity_log[date] = activity

    def set_goal(self, goal):
        self.goal = goal

    def generate_report(self, date):
        if date in self.activity_log:
            activity = self.activity_log[date]
            self.goal.progress_report(activity)
        else:
            print("No activity logged for this date.")

# Example Usage:
# Create a user
user = User("Alice", 30, "Female", 60, 165)

# Log activity
activity1 = Activity("2024-02-01", steps=5000, workouts=["Running", "Yoga"], calorie_intake=1800, sleep_hours=7)
user.log_activity("2024-02-01", activity1)

# Set goal
goal = Goal(target_steps=10000, target_workouts_per_week=3, target_calories_per_day=2000, target_sleep_hours_per_night=8)
user.set_goal(goal)

# Generate report
user.generate_report("2024-02-01")
```

```python
'''**9. Music Player:**

- Create classes for `Song`, `Album`, `Playlist`, and `Player`.
- Organize music library, create playlists, and shuffle songs.
- Control playback, adjust volume, and set sound settings.
- Display lyrics and visualize audio.
'''
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing '{self.title}' by {self.artist}")

class Album:
    def __init__(self, title, artist, songs):
        self.title = title
        self.artist = artist
        self.songs = songs

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def remove_song(self, song):
        self.songs.remove(song)

    def shuffle(self):
        random.shuffle(self.songs)

class Player:
    def __init__(self):
        self.volume = 50
        self.playing = False

    def play_song(self, song):
        song.play()
        self.playing = True

    def pause(self):
        self.playing = False

    def adjust_volume(self, volume):
        self.volume = volume

# Example Usage:
# Create songs
song1 = Song("Song 1", "Artist 1", "3:45")
song2 = Song("Song 2", "Artist 2", "4:20")
song3 = Song("Song 3", "Artist 3", "3:15")

# Create playlist
playlist = Playlist("My Playlist")
playlist.add_song(song1)
playlist.add_song(song2)
playlist.add_song(song3)

# Create player
player = Player()

# Play playlist
for song in playlist.songs:
    player.play_song(song)

# Pause
player.pause()

# Adjust volume
player.adjust_volume(70)
```

```python
'''
**10. Social Media Platform:**

- Create classes for `User`, `Post`, `Comment`, and `Message`.
- Implement features for creating profiles,'''
class User:
    def __init__(self):
        self.username = input("Enter name: ")
        self.age = input("Enter age: ")
        self.followers = input("Enter followers: ")

    def display_user_info(self):
        print(f"Username: {self.username}, Age: {self.age}, Followers: {self.followers}")


class Post:
    def __init__(self, author, content, likes=0, comments=[]):
        self.author = author
        self.content = content
        self.likes = likes
        self.comments = comments

    def display_post_info(self):
        print(f"Author: {self.author}, Content: {self.content}, Likes: {self.likes}\nComments: {self.comments}")


class SocialMediaAnalysisTool:
    def __init__(self):
        self.users = []
        self.posts = []

    def add_user(self, user):
        self.users.append(user)

    def add_post(self, post):
        self.posts.append(post)

    def display_user_stats(self):
        print(f"User Statistics: ")
        for user in self.users:
            user.display_user_info()

    def display_post_stats(self):
        print(f"Post Statistics: ")
        for post in self.posts:
            post.display_post_info()

    def analyze_engagement(self):
        print("Analyze Report:")
        total_likes = sum(post.likes for post in self.posts)
        total_comments = sum(len(post.comments) for post in self.posts)
        average_likes_per_post = total_likes / len(self.posts) if len(self.posts) > 0 else 0
        average_comments_per_post = total_comments / len(self.posts) if len(self.posts) > 0 else 0

        print(f"Total Likes: {total_likes}")
        print(f"Total Comments: {total_comments}")
        print(f"Average Likes per Post: {average_likes_per_post}")
        print(f"Average Comments per Post: {average_comments_per_post}")


if __name__ == "__main__":
    # Creating Users:
    user1 = User()

    # Creating Posts:
    post1 = Post(user1.username, "Excited about my new project!", likes=50, comments=["Great job!", "Looking forward to it"])
    post2 = Post(user1.username, "Excited about my new project!", likes=70, comments=["Great job!", "Looking forward"])

    # Creating social analysis tool
    analysis_tool = SocialMediaAnalysisTool()

    # Adding users and posts to the Analyze Tool
    analysis_tool.add_user(user1)
    analysis_tool.add_post(post1)
    analysis_tool.add_post(post2)

    # Displaying user and post statistics
    analysis_tool.display_user_stats()
    print("\n")
    analysis_tool.display_post_stats()

    # Analyzing Engagement
    analysis_tool.analyze_engagement()

```
